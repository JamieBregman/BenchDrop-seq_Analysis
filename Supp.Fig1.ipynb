```python
# Import
import integrative_transcriptomics_viewer as itv
from integrative_transcriptomics_viewer.convenience import Configuration
from integrative_transcriptomics_viewer.export import save
from integrative_transcriptomics_viewer.classification import BAMtagClassification
from integrative_transcriptomics_viewer.annotation_matching import TaggedBAMAnnotationMatching
```

```python
# Reference files 
human_genome_path = "genome.fa"
human_gtf_path = "genes.gtf"
human_bed_gencode = "genes.bed.gz"

human = Configuration(genome_fasta = human_genome_path, gtf_annotation = human_gtf_path, bed_annotation = [str(human_bed_gencode)])
```

```python
# Input BAM file
all_bams_dict = {"Short":"SR.K562.sort.bam",
                "BenchDrop-seq":"LR.K562.genome.sort.bam"} 
```

```python
gene = "HBZ"
```

```python
# View in notebook

human.plot_feature(gene, 
    bams_dict = all_bams_dict, 
    # with_reads = False,
    with_reads = True,
    tighter_track = True,
    add_track_label = gene,
    add_coverage_label = False,
    vertical_layout_reads = True)
```

```python
# Save plot

plot = human.plot_feature(gene, 
    bams_dict = all_bams_dict, 
    # with_reads = False,
    with_reads = True,
    tighter_track = True,
    add_track_label = gene,
    add_coverage_label = False,
    vertical_layout_reads = True,
    )

save(plot, f"{gene}.svg")
```
